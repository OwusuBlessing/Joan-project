```python
import requests
from bs4 import BeautifulSoup

def get_html_page(url):
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception if the request was unsuccessful
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    return soup
    



```

```python
url = """https://jiji.ng/motorcycles-and-scooters?query=car%20oil%20filters"""

text = get_html_page(url)
```

```python
price_elements = text.find_all("div", class_="qa-advert-price")
```

```python
parent_class = "b-list-advert-base__data__header"
price_elements = text.find_all(class_=parent_class)

details = []

for element in price_elements:
    price_element = element.find(class_="qa-advert-price")
    title_element = element.find("div", class_="b-advert-title-inner qa-advert-title b-advert-title-inner--div")

    if price_element and title_element:
        price = price_element.text.strip()
        title = title_element.text.strip()
        dict_ = {"name": title, "price": price}
        details.append(dict_)
    

details
```
